### 🔗 Fluxo de Dados
Banco Relacional ➝ Limpeza e Padronização ➝ Modelo Dimensional ➝ Dashboard

```python
### **3. Conexão com o Banco e Pré-processamento**

import pandas as pd
import psycopg2
from sqlalchemy import create_engine

# Exemplo de conexão (ajuste para suas credenciais)
engine = create_engine('postgresql://usuario:senha@localhost:5432/analytics_commerce')

# Carregar tabela de vendas brutas
df_vendas = pd.read_sql("SELECT * FROM vendas", engine)
df_vendas.head()
```

```python
### **4. Limpeza e Padronização de Dados**
import re

def limpar_nome(nome):
    nome = nome.strip().upper()
    nome = re.sub(r'Á|À|Â|Ã', 'A', nome)
    nome = re.sub(r'É|Ê', 'E', nome)
    nome = re.sub(r'Í', 'I', nome)
    nome = re.sub(r'Ó|Ô|Õ', 'O', nome)
    nome = re.sub(r'Ú|Ü', 'U', nome)
    nome = re.sub(r'Ç', 'C', nome)
    nome = re.sub(r'[^A-Z ]', '', nome)
    return nome

df_vendas['cliente_nome_limpo'] = df_vendas['cliente_nome'].apply(limpar_nome)
df_vendas[['cliente_nome', 'cliente_nome_limpo']].head()
```

## ⭐ Construção do Modelo Dimensional

As seguintes views foram criadas no PostgreSQL:
- `dim_clientes`
- `dim_vendedores`
- `dim_produtos`
- `dim_tempo`
- `fato_vendas`

A modelagem segue o **padrão Star Schema**, com `fato_vendas` centralizando as métricas.



```python
### **6. KPIs e Métricas**

# Exemplo: Ticket Médio
df_ticket = pd.read_sql("""
    SELECT c.nome AS cliente, SUM(v.valor_total)/COUNT(DISTINCT v.id) AS ticket_medio
    FROM fato_vendas v
    JOIN dim_clientes c ON v.cliente_id = c.id
    GROUP BY c.nome
    ORDER BY ticket_medio DESC
    LIMIT 10;
""", engine)
df_ticket
```
